```python
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns


rango= np.random.RandomState(0)
x=np.linspace(0,10,500)
y=np.cumsum(rango.randn(500,6),0)

#Graficar los datos
plt.plot(x,y)
plt.legend('ABCFDE',ncol=2,loc='upper left');


#Graficar los datos
sns.set()#carga de estilos de seaborn
plt.plot(x,y)
plt.legend('ABCFDE',ncol=2,loc='upper left');


```

```python
#generando puntos de una funcin

x_val=np.linspace(0,10,20)
y_val= x_val**2
plt.scatter(x_val, y_val)
```

```python

plt.scatter(x_val, y_val)
plt.title("Desplazamiento")
plt.xlabel("tiempo[segundos]")
plt.ylabel("Distancia[metros]")
```

```python
plt.scatter(x_val, y_val, marker='*', color='r')
plt.title("Desplazamiento")
plt.xlabel("tiempo[segundos]")
plt.ylabel("Distancia[metros]")
```

```python
plt.plot(x_val, y_val, color='r')
plt.scatter(x_val, y_val, marker=',', color='g', s=150)
plt.title("Desplazamiento")
plt.xlabel("tiempo[segundos]")
plt.ylabel("Distancia[metros]")
```
